```python
base_dir = "gs://bimpredict-data-bucket"
dfs = []

for i in range(1, 7):
    murs_path = os.path.join(base_dir, f"maquette{i}", f"murs{i}.csv")

    name = f"maquette{i}/murs{i}.csv"
    storage_client = storage.Client()
    bucket_name = 'bimpredict-data-bucket'
    bucket = storage_client.bucket(bucket_name)
    state = storage.Blob(bucket=bucket, name=name).exists(storage_client)

    if state:
        dfs.append(pd.read_csv(murs_path, sep=';'))
    else:
        print(f"Fichier non trouvé : {murs_path}")
```
